```python
alphabets = ['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z','\n']
one = list(np.zeros((len(alphabets))))
alpha_dict={}
for n,i in enumerate(alphabets):
    one[n]=1
    alpha_dict[i] = one
    one = list(np.zeros((len(alphabets))))
print(alpha_dict)
```

```python
#my
class RNN(nn.Module):
    def __init__(self,inputsize,hiddensize,outputsize):
        super(RNN,self).__init__()
        self.inptoconc = nn.Linear(inputsize,inputsize)
        self.conctoop = nn.Linear(hiddensize, outputsize)
        self.conctohidden = nn.Linear(inputsize+hiddensize, hiddensize)
        self.tanh = nn.Tanh()
    def inpnn(self,x):
        return self.inptoconc(x)
    def forward(self,combined):
        hid = self.conctohidden(combined)
        hid = self.tanh(hid)
        out = self.conctoop(hid)
        return out,hid
```

```python
model = RNN(27,50,27)
print(model)
```

```python
learning_rate = 0.0001
h0 = torch.Tensor(np.zeros((50)))
lossfn = nn.CrossEntropyLoss()
epoch_loss_list_avg = []
epoch_loss_list = []
for epochs in range(50):
    for i in name_dict:
        word_loss = 0
        count = 0
        for j in range(len(name_dict[i])-1):
            model.zero_grad()
            if count == 0:
                x = model.inpnn(name_dict[i][j])
                a = torch.cat((x.view(-1),h0.view(-1)))
                pred,h=model.forward(a)
            else:
                x = model.inpnn(name_dict[i][j])
                a = torch.cat((x.view(-1),h.view(-1)))
                pred,h=model.forward(a)
            count+=1
            loss = lossfn(pred.view(1,-1),torch.Tensor([target[i][j+1]]).type(torch.LongTensor))
            loss.backward(retain_graph=True)
            word_loss+=loss.item()
            for w in model.parameters():
                w.data = w.data - learning_rate*w.grad.data
        print('.',end='')
    epoch_loss_list.append(word_loss)
    word_loss/=(len(i)+1)
    epoch_loss_list_avg.append(word_loss)
    print()
    print('average loss on epoch ',epochs+1,'= ',sum(epoch_loss_list_avg)/len(epoch_loss_list_avg))
    print('loss on epoch ',epochs+1,'= ',sum(epoch_loss_list)/len(epoch_loss_list))
    
```
